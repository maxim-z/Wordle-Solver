```python
df = pd.read_csv("frequency_list_for_heatmap.csv")
# drop the first column as its just the alphabet
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
x_axis_labels = ['Spot 1', 'Spot 2', 'Spot 3', 'Spot 4', 'Spot 5']
y_axis_labels = ['a', 'b', 'c', 'd', 'f', 'e', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
df.set_axis(y_axis_labels, axis='index')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5,6.5))         # Sample figsize in inches
freq_heatmap = sns.heatmap(df.iloc[:, :6:], annot=False, linewidths=.5, ax=ax, cmap="YlGnBu", 
                           xticklabels=x_axis_labels, yticklabels=y_axis_labels)
fig.savefig('freq_heatmap.png')
```
